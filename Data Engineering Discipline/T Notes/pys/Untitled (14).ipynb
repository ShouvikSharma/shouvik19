```pyspark
import boto3
from pyspark.sql import SparkSession
from pyspark.sql.functions import lit, col
from pyspark.sql.utils import AnalysisException
from pyspark.sql.types import StructType, StructField, StringType, IntegerType
import gc
from concurrent.futures import ThreadPoolExecutor
from typing import List

def create_optimized_spark_session() -> SparkSession:
    """Create an optimized Spark session with tuned configurations."""
    return SparkSession.builder \
        .appName("CombineParquetFiles") \
        .config("spark.driver.memory", "4g") \
        .config("spark.executor.memory", "8g") \
        .config("spark.sql.files.maxPartitionBytes", "128MB") \
        .config("spark.sql.shuffle.partitions", "200") \
        .config("spark.memory.offHeap.enabled", "true") \
        .config("spark.memory.offHeap.size", "2g") \
        .config("spark.sql.adaptive.enabled", "true") \
        .config("spark.sql.adaptive.coalescePartitions.enabled", "true") \
        .config("spark.sql.adaptive.skewJoin.enabled", "true") \
        .config("spark.default.parallelism", "100") \
        .config("spark.serializer", "org.apache.spark.serializer.KryoSerializer") \
        .config("spark.sql.parquet.compression.codec", "snappy") \
        .getOrCreate()

def list_s3_folders(s3_client: boto3.client, bucket: str, prefix: str) -> List[str]:
    """List S3 folders using pagination with improved error handling."""
    try:
        paginator = s3_client.get_paginator("list_objects_v2")
        folders = []
        for page in paginator.paginate(Bucket=bucket, Prefix=prefix, Delimiter="/"):
            if "CommonPrefixes" in page:
                folders.extend([obj["Prefix"].split("/")[-2] 
                              for obj in page["CommonPrefixes"]])
        return folders
    except Exception as e:
        print(f"Error listing S3 folders: {str(e)}")
        raise

def align_to_schema(df, schema: StructType):
    """Align DataFrame to schema with optimized column handling."""
    missing_cols = set(field.name for field in schema.fields) - set(df.columns)
    for col_name in missing_cols:
        field = next(f for f in schema.fields if f.name == col_name)
        df = df.withColumn(col_name, lit(None).cast(field.dataType))
    return df.select(*[col(field.name) for field in schema.fields])

def process_folder(spark: SparkSession, bucket: str, prefix: str, 
                  folder: str, output_path: str, schema: StructType) -> None:
    """Process a single folder with optimized reading and writing."""
    input_path = f"s3://{bucket}/{prefix}{folder}/*.pq"
    output_folder_path = f"{output_path}{folder}/"
    
    try:
        # Read with optimized configurations
        df = spark.read \
            .option("mergeSchema", "true") \
            .option("vectorizedReader", "true") \
            .option("spark.sql.parquet.filterPushdown", "true") \
            .parquet(input_path)
        
        # Calculate optimal partition count based on data size
        total_size_bytes = sum(f.size for f in df._jdf.inputFiles())
        partition_size = 128 * 1024 * 1024  # 128MB per partition
        num_partitions = max(1, total_size_bytes // partition_size)
        
        # Process with optimized operations
        df_aligned = align_to_schema(df, schema)
        df_final = df_aligned.withColumn("source_folder", lit(folder)) \
            .repartition(num_partitions)
        
        # Write with optimized configurations
        df_final.write \
            .option("compression", "snappy") \
            .option("spark.sql.parquet.writeLegacyFormat", "false") \
            .mode("overwrite") \
            .parquet(output_folder_path)
        
        print(f"Successfully processed folder: {folder}")
        
        # Clean up
        df_final.unpersist()
        spark.catalog.clearCache()
        del df, df_aligned, df_final
        gc.collect()
        
    except Exception as e:
        print(f"Error processing folder {folder}: {str(e)}")
        raise

def main():
    # Configuration
    bucket_name = "avantcamp-partner01-prod-ingestion-bucket"
    prefix = "avant/direct_mail/mailed/"
    output_path = "s3://avantcamp-partner01-prod-ingestion-bucket/avant/direct_mail/combined_mailed_data/"
    batch_size = 5  # Reduced batch size for better memory management
    
    # Initialize services
    spark = create_optimized_spark_session()
    s3 = boto3.client("s3")
    
    try:
        # Get folders
        folders = list_s3_folders(s3, bucket_name, prefix)
        if not folders:
            print("No folders found under the specified prefix.")
            return
        
        print(f"Found {len(folders)} folders to process")
        
        # Process folders in batches
        for i in range(0, len(folders), batch_size):
            batch = folders[i:i + batch_size]
            print(f"Processing batch {i//batch_size + 1}/{len(folders)//batch_size + 1}")
            
            # Process batch in parallel
            with ThreadPoolExecutor(max_workers=batch_size) as executor:
                futures = [
                    executor.submit(
                        process_folder, 
                        spark, 
                        bucket_name, 
                        prefix, 
                        folder, 
                        output_path, 
                        schema
                    )
                    for folder in batch
                ]
                
                # Wait for batch completion
                for future in futures:
                    future.result()
            
            # Force cleanup between batches
            spark.catalog.clearCache()
            gc.collect()
            
    except Exception as e:
        print(f"Error in main process: {str(e)}")
        raise
    finally:
        spark.stop()

if __name__ == "__main__":
    main()
```
